```python
import numpy as np
from numpy import random
from scipy import stats
```

### One-sample T-test

```python
from scipy.stats import ttest_1samp
```

```python
ages = [10,20,35,50,28,40,55,18,16,55,30,25,43,18,30,28,14,24,16,17,32,35,26,27,65,18,43,23,21,20,19,70]
len(ages)
```

```python
age_mean = np.mean(ages)
print(age_mean)
```

```python
sample_size = 10
sample = random.choice(ages, sample_size) #random.choice(arr, size)
print(sample)
```

```python
ttest, p_val = ttest_1samp(sample, 30)  #ttest_1samp(sample, expected_val)
print(p_val)
```

```python
#H0 - no difference between sample and population mean
#H1 - difference between sample and poplulation mean
if p_val<0.05:
    print('reject null hypothesis')
else:
    print('accept null hypothesis')
```

### Independent samples T-Test

```python
random.seed(12)
classA = stats.poisson.rvs(loc=18, mu=30, size=60)
classB = stats.poisson.rvs(loc=18, mu=33, size=60)
print('classA mean:',classA.mean())
print('classB mean:', classB.mean())
```

```python
_, p_val = stats.ttest_ind(a=classA, b=classB, equal_var=False)
print(p_val)
```

```python
#H0 - population means are same
#H1 - populations are different
if p_val<0.05:
    print('reject null hypothesis')
else:
    print('accept null hypothesis')
```

### Paired two-sample T-test

```python
math1 = np.array([4, 4, 7, 16, 20, 11, 13, 9, 11, 15])
math2 = np.array([15, 16, 14, 14, 22, 22, 23, 18, 18, 19])
```

```python
_,p_val = stats.ttest_rel(a=math1, b=math2)
print(p_val)
```

```python
if p_val<0.05:
    print('reject null hypothesis')
else:
    print('accept null hypothesis')
```

### Chi-square Test

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df_table = pd.crosstab(df['sex'],df['smoker'])
print(df_table)
```

```python
observed_vals = df_table.values
print('Observed:\n', observed_vals)
```

```python
val = stats.chi2_contingency(df_table)
print(val) #returns test-statistic, p-val, dof, expected vals
```

```python
expected_vals = val[3]
expected_vals
```

```python
r = len(df_table.iloc[0:2, 0])
c = len(df_table.iloc[0, 0:2])
ddof = (r-1)*(c-1)
print('Degree of freedom:',ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_sq = sum([(o-e)**2/e for o,e in zip(observed_vals, expected_vals)])
chi_sq_stat = chi_sq[0]+chi_sq[1]
print('Chi-Square Statistic:',chi_sq_stat)
```

```python
critical_val = chi2.ppf(q=1-alpha, df=ddof)
print('Critical value:',critical_val)
```

```python
if chi_sq_stat >= critical_val:
    print('Reject H0, there is relationship between the categorical variables')
else:
    print('Fail to reject H0, no significant relationship')
```

### One-Way ANOVA 

```python
from scipy.stats import f_oneway
```

```python
p1 = [89, 89, 88, 78, 79]
p2 = [93, 92, 94, 89, 88]
p3 = [89, 88, 89, 93, 90]
p4 = [81, 78, 81, 92, 82]
```

```python
f_stat, p_value = f_oneway(p1, p2, p3, p4)
print(f'F-statistic value:{f_stat}')
print(f'p-value:{p_value}')
```

```python
if p_value<0.05:
    print('Reject the null hypothesis')
else:
    print('Fail to reject null hypothesis')
```

### One-sample Z-Test

```python
from statsmodels.stats.weightstats import ztest
```

```python
data = [11.8]*100
pop_mean = 12
pop_sd = 0.5

z_stat, p_val = ztest(data, value=pop_mean)
print(f'Z-statistic:{z_stat}')
print(f'p-value:{p_val}')
```

```python
alpha = 0.05
if p_val<alpha:
    print('reject null hypothesis, avg battey life different from claimed')
else:
    print('fail to reject null hypothesis, avg battery life equal to claimed')
```

### F-Test

```python
grp1 = np.random.rand(25)
grp2 = np.random.rand(20)

var1 = np.var(grp1, ddof=1)
var2 = np.var(grp2, ddof=1)

f_val = var1/var2

df1 = len(grp1)-1
df2 = len(grp2)-1

p_val = 2*min(stats.f.cdf(f_val, df1, df2), 1 - stats.f.cdf(f_val, df1, df2))

print('dof 1:', df1)
print('dof 2:',df2)
print('F-statistic:', f_val)
print('p-value:',p_val)
```

```python
alpha = 0.05
if p_val<alpha:
    print('reject null hypothesis, variances are different')
else:
    print('fail to reject null hypothesis, no significant difference')
```
