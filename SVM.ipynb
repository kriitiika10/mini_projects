```python
df0 = df[df['Species']=='Iris-setosa']
df1 = df[df['Species'] == 'Iris-versicolor']
df2 = df[df['Species'] == 'Iris-virginica']
```

```python
df.drop('Id', axis=1, inplace=True)
df.head()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:,-1]
```

```python
y = y.map({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred, y_test)
```

```python
from sklearn.datasets import load_digits
dig = load_digits()
```

```python
df = pd.DataFrame(dig.data, columns=dig.feature_names)
df.head()
```

```python
df['target'] = dig.target
df.head()
```

```python
df.shape
```

```python
plt.matshow(dig.images[3], cmap='gray')
plt.show()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
parameters = {
    'C': [0.1,1,10,20],
    'kernel': ['rbf','linear','poly','sigmoid']
}
```

```python
s1 = SVC(C=0.1, kernel='poly', degree=3, gamma='scale')
```

```python
s1.fit(X_train,y_train)
y_pred = s1.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
s2 = SVC(C=1, kernel='poly', degree=3, gamma='scale')
s2.fit(X_train, y_train)
y_pred = s2.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
s3 = SVC(C=10, kernel='rbf', gamma='scale')
s3.fit(X_train, y_train)
y_pred = s3.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
s4 = SVC(C=10, kernel='linear', gamma='scale')
s4.fit(X_train, y_train)
y_pred = s4.predict(X_test)
accuracy_score(y_pred, y_test)
```
